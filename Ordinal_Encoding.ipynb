# **Ordinal Encoding Example**

# **1 Import necessary libraries**


```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

# **2 Create Sample Data**

1.  Let's create a DataFrame with educational qualifications,
2. which is a common categorical variable where order matters.
3.  We'll use a sample of qualifications in a typical Indian context.
4.  The order `(10th Pass < 12th Pass < Undergraduate < Postgraduate < Doctorate)` is inherent.

```python
# Sample data representing educational qualifications

data = {
    'Student_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Qualification': [
        '10th Pass', 'Undergraduate', 'Postgraduate', '12th Pass', 'Doctorate',
        'Undergraduate', '10th Pass', 'Postgraduate', '12th Pass', 'Undergraduate'
    ]
}

df = pd.DataFrame(data)
display(df)
```

# **3 Define the Ordinal Order**
1. When using `OrdinalEncoder`, it's crucial to explicitly define the order of categories if there's a natural ranking.
2. This ensures that the encoding is done correctly according to the inherent order.
3. Here, we define the order for 'Qualification' from lowest to highest educational level.

```python
# Define the specific order for the 'Qualification' column

qualification_order = [
    '10th Pass',
    '12th Pass',
    'Undergraduate',
    'Postgraduate',
    'Doctorate'
]
```

*  Initialize the OrdinalEncoder with the specified categories order
* We pass a list of lists because OrdinalEncoder expects a list of categories for each feature

```python
encoder = OrdinalEncoder(
    categories=[qualification_order]
    )
```

# **4  Fit and transform the 'Qualification' column**


```python
df['Qualification_Encoded'] = encoder.fit_transform(df[['Qualification']])
```

```python
display(df)
```

# Explanation

- **`Qualification_Encoded`**: This new column contains the numerical representation of the 'Qualification' column based on the order we defined:
    - `10th Pass` is encoded as `0.0`
    - `12th Pass` is encoded as `1.0`
    - `Undergraduate` is encoded as `2.0`
    - `Postgraduate` is encoded as `3.0`
    - `Doctorate` is encoded as `4.0`

This method is suitable when the categories have a clear,

 meaningful order, and you want to preserve that order in the numerical representation for machine learning models.
