# **One-Hot Encoding with Indian State Names (using Scikit-learn)**

# **1 Install the Library**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

# **2 Create a sample DataFrame with Indian state names**

```python
indian_states_data = {
    'Customer_ID': [101, 102, 103, 104, 105, 106],
    'State': ['Maharashtra', 'Delhi', 'Karnataka', 'Maharashtra', 'Uttar Pradesh', 'Tamil Nadu'],
    'Purchase_Amount': [1500, 800, 2200, 1700, 1000, 3000]
}

df_indian_states = pd.DataFrame(indian_states_data)
```

# **3 Display the DataFrame**

```python
display(df_indian_states)
```

# **4 Initialize the OneHotEncoder**
* sparse_output=False ensures a dense array output

```python
state_encoder = OneHotEncoder(
    handle_unknown='ignore',
    sparse_output=False
                          )
```

# **5 Fit and transform the 'State' column**
* Reshape to 2D array as scikit-learn transformers expect

```python
states_encoded = state_encoder.fit_transform(df_indian_states[['State']])
```

```python
states_encoded
```

# **6 Create a DataFrame from the encoded data**


```python
states_encoded_df = pd.DataFrame(
    states_encoded,
    columns=state_encoder.get_feature_names_out(['State'])
    )
```

```python
states_encoded_df
```

# **7 Concatenate with the original DataFrame (dropping the original 'State' column)**


```python
df_indian_states_encoded = pd.concat(
    [df_indian_states.drop(columns=['State']),
     states_encoded_df], axis=1
    )
```
